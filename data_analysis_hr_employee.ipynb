```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Load Dataset

df=pd.read_csv('HR-Employee.csv')
```

```python
# Create folder for saving plots
os.makedirs('images', exist_ok=True)
```

```python
# --- 1. Basic Info & Summary ---
print("=== Dataset Info ===")
print(df.info())
print("\n=== Numeric Summary ===")
print(df.describe())
print("\n=== Categorical Summary ===")
print(df.describe(include='object'))
```

```python
# --- 2. Missing Values ---
print("\n=== Missing Values ===")
print(df.isnull().sum())
```

```python
# --- 3. Numeric Features Distribution ---
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=20, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.savefig(f'images/dist_{col}.png')
    plt.close()
```

```python
# --- Categorical Features Count Plots ---
categorical_cols = df.select_dtypes(include='object').columns
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(
        data=df, 
        x=col, 
        order=df[col].value_counts().index, 
        color='skyblue'  # use a single color instead of palette
    )
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f'images/count_{col}.png')
    plt.close()
```

```python
# --- 5. Correlation Heatmap ---
plt.figure(figsize=(12,10))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('images/correlation_heatmap.png')
plt.close()
```
